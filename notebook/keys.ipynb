```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def create_keys(keys: dict) -> pd.DataFrame:
    df = pd.DataFrame(keys)
    df['press_time'] = df['ups'] - keys['downs']
    return df
```

```python
firefox = create_keys({
    "ups":[144,194,393,499,717,845],
    "downs":[0,69,266,383,593,746]
})
chrome = create_keys({
    "ups":[170,243,404,502,755,874],
    "downs":[0,129,293,372,611,787]
})
other_keyboard = create_keys({
    "ups":[134,222,390,518,694,830],
    "downs":[0,151,294,422,591,750]
})
```

```python
fig, ax = plt.subplots()
ax.plot(firefox['downs'], firefox['press_time'], label="firefox")
ax.plot(chrome['downs'], chrome['press_time'], label="chrome")
ax.plot(other_keyboard['downs'], other_keyboard['press_time'], label="other_keyboard")

ax.set_title("Key signature")
ax.legend()
ax.set_xlabel("key down over time")
ax.set_ylabel("key press time")
plt.show()
```

```python
first = create_keys({
    "ups":[94,233,271,334,405,504],
    "downs":[0, 57, 134, 187, 267, 366]
})
second = create_keys({
    "ups":[111,274,307,383, 427, 523],
    "downs":[0, 78,177, 235, 320, 399]
})
third = create_keys({
    "ups":[137, 246, 293, 406, 428, 547],
    "downs":[0, 28, 146, 223, 286, 376]
})
mean = {"ups":[], "downs": []}
for i in range(6):
    up  = first.at[i, "ups"] +  second.at[i, "ups"] + third.at[i, "ups"]
    down  = first.at[i, "downs"] +  second.at[i, "downs"] + third.at[i, "downs"]
    mean["ups"].append(up/3)
    mean["downs"].append(down/3)
mean = create_keys(mean)
    
```

```python
fig, ax = plt.subplots()
plot = lambda df, label: ax.plot(df['downs'], df['press_time'], label=label)
plot(first, "1")
plot(second, "2")
plot(third, "3")
ax.plot(mean['downs'], mean['press_time'], '--', label="mean")
ax.set_title("Key signature")
ax.legend()
ax.set_xlabel("key down over time")
ax.set_ylabel("key press time")
plt.show()
```
